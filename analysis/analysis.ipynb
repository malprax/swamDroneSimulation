# SwarmDroneSim Analysis
Notebook ini membaca `detections.csv` dan membuat grafik:
- Histogram waktu deteksi
- Boxplot
- Tabel ringkas

```python
import pandas as pd
import matplotlib.pyplot as plt
import os
csv_path = 'detections.csv'  # salin dari user://logs/detections.csv
assert os.path.exists(csv_path), 'Letakkan detections.csv di folder yang sama dengan notebook ini.'
df = pd.read_csv(csv_path)
df.head()
```

```python
df.describe(include='all')
```

```python
plt.figure()
df['time_sec'].hist(bins=20)
plt.title('Histogram Detection Time')
plt.xlabel('seconds'); plt.ylabel('count')
plt.show()
```

```python
plt.figure()
plt.boxplot(df['time_sec'], vert=True, labels=['det time'])
plt.title('Boxplot Detection Time')
plt.ylabel('seconds')
plt.show()
```
